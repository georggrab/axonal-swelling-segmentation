```python
import tifffile
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

sns.set_theme(style="whitegrid", context="paper")

THRESHOLDS = [.50, .80, .95]
arr = tifffile.imread("out2.tif")
```

```python
for thresh in THRESHOLDS:
    arr_thresh = (arr > thresh).reshape(-1, arr.shape[1] * arr.shape[2])
    arr_counts = np.count_nonzero(arr_thresh, axis=1)
    plt.plot(arr_counts, label=f"$p={thresh}$")
plt.legend()
plt.xlabel("frame")
plt.ylabel("number of pixels detected as swellings");

plt.savefig("output_figures/detections_naive.pdf")
```
